```python
import os
import requests
from langchain import LLMChain
from langchain.prompts import PromptTemplate
from typing import ClassVar
from dotenv import load_dotenv
```

```python
load_dotenv()  # This loads environment variables from the .env file
api_key = os.getenv("HUGGING_FACE_API_KEY")

# Set your Hugging Face API key
os.environ['HUGGING_FACE_API_KEY'] = 'Place_your_own_API_key'

# Create a custom chain to handle text generation logic
class HuggingFaceTextGenerationChain(LLMChain):
    API_URL: ClassVar[str] = "https://api-inference.huggingface.co/models/gpt2"

    def __init__(self):
        # You can define a prompt template here if needed, adjust as appropriate
        prompt_template = PromptTemplate(input_variables=["prompt"], template="{prompt}")
        super().__init__(prompt=prompt_template, llm=self)

    def _call(self, inputs):
        """Override the call method to send requests to the Hugging Face API."""
        headers = {"Authorization": f"Bearer {os.getenv('HUGGING_FACE_API_KEY')}"}
        payload = {"inputs": inputs["prompt"]}
        response = requests.post(self.API_URL, headers=headers, json=payload)

        if response.status_code == 200:
            return response.json()[0]['generated_text']
        else:
            return f"Error: {response.status_code} - {response.text}"

    def run(self, prompt):
        """Process the input and get a result."""
        return self._call({"prompt": prompt})

def main():
    text_chain = HuggingFaceTextGenerationChain()
    while True:
        user_input = input("Enter your prompt (or 'exit' to quit): ").strip()
        if user_input.lower() == "exit":
            break
        result = text_chain.run(user_input)
        print("Generated Text:\n", result)
        print("\n" + "-"*50 + "\n")

if __name__ == "__main__":
    main()
```
